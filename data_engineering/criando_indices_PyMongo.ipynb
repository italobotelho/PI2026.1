###<font color = "#5ccda7">**1. ГЌNDICES - casos geolocalizados**

```python
db.casos_geolocalizados.create_index(
    [("DT_NOTIFIC", pymongo.ASCENDING)],
    name="idx_casos_data_notificacao"
)

db.casos_geolocalizados.create_index(
    [("ID_UNIDADE", pymongo.ASCENDING)],
    name="idx_casos_id_unidade"
)

db.casos_geolocalizados.create_index(
    [("NO_FANTASIA", pymongo.ASCENDING)],
    name="idx_casos_nome_hospital"
)

db.casos_geolocalizados.create_index(
    [("CS_SEXO", pymongo.ASCENDING)],
    name="idx_casos_sexo"
)

db.casos_geolocalizados.create_index(
    [("NU_IDADE_N", pymongo.ASCENDING)],
    name="idx_casos_idade"
)

db.casos_geolocalizados.create_index(
    [("HOSPITALIZ", pymongo.ASCENDING)],
    name="idx_casos_hospitalizacao"
)

db.casos_geolocalizados.create_index(
    [("ID_MUNICIPIO", pymongo.ASCENDING)],
    name="idx_casos_municipio"
)

db.casos_geolocalizados.create_index(
    [("DT_NOTIFIC", pymongo.ASCENDING), ("ID_UNIDADE", pymongo.ASCENDING)],
    name="idx_casos_data_unidade"
)

db.casos_geolocalizados.create_index(
    [("LATITUDE", pymongo.ASCENDING), ("LONGITUDE", pymongo.ASCENDING)],
    name="idx_casos_lat_long"
)
```

```python
db.casos_geolocalizados.create_index(
    [("NOME_DOENCA", pymongo.ASCENDING)],
    name="idx_casos_doenca"
)

db.casos_geolocalizados.create_index(
    [("DT_NOTIFIC", pymongo.ASCENDING), ("NOME_DOENCA", pymongo.ASCENDING)],
    name="idx_casos_data_doenca"
)
```

###<font color = "#5ccda7"> **2. ГЌNDICES - dados climГЎticos**

```python
db.clima.create_index(
    [("DT_MEDICAO", pymongo.ASCENDING)],
    name="idx_clima_data_medicao"
)

db.clima.create_index(
    [("PRECIP_TOTAL", pymongo.DESCENDING)],
    name="idx_clima_precipitacao"
)

db.clima.create_index(
    [("TEMP_MED", pymongo.DESCENDING)],
    name="idx_clima_temperatura"
)

db.clima.create_index(
    [("DT_MEDICAO", pymongo.ASCENDING), ("PRECIP_TOTAL", pymongo.DESCENDING)],
    name="idx_clima_data_precipitacao"
)

db.clima.create_index(
    [("DT_MEDICAO", pymongo.ASCENDING), ("TEMP_MED", pymongo.DESCENDING)],
    name="idx_clima_data_temperatura"
)
```

### <font color = "#5ccda7">**3. ГЌNDICES - risco de inundaГ§ГЈo**

```python
db.risco_inundacao.create_index(
    [("properties.CLASSE", pymongo.ASCENDING)],
    name="idx_risco_classe"
)

db.risco_inundacao.create_index(
    [("geometry", pymongo.GEOSPHERE)],
    name="idx_risco_geo"
)
```

### <font color = "#5ccda7">**4. ГЌNDICES - vunerabilidade habitacional**

```python
db.vulnerabilidade_habitacional.create_index(
    [("properties.NOME_AREA", pymongo.ASCENDING)],
    name="idx_vulnerabilidade_nome"
)

db.vulnerabilidade_habitacional.create_index(
    [("geometry", pymongo.GEOSPHERE)],
    name="idx_vulnerabilidade_geo"
)
```
