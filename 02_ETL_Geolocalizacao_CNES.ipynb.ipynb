```python
import pandas as pd
import os
```

```python
from pysus.ftp.databases.cnes import CNES

print("Baixando amostra do CNES para inspeção...")
cnes = CNES().load()
arquivos_cnes = cnes.get_files('ST', year=2025, month=12, uf='SP')
objetos_cnes = cnes.download(arquivos_cnes)

# Pega apenas o primeiro objeto retornado para ver a estrutura
if isinstance(objetos_cnes, list):
    df_amostra = objetos_cnes[0].to_dataframe()
else:
    df_amostra = objetos_cnes.to_dataframe()

print("\nAs colunas reais e oficiais enviadas pelo DATASUS são:")
print(list(df_amostra.columns))
```

```python
COL_MUNICIPIO = 'CO_IBGE'
COL_CNES = 'CO_CNES'
COL_NOME = 'NO_FANTASIA'
COL_LAT = 'NU_LATITUDE'
COL_LON = 'NU_LONGITUDE'

# ---------------------------------------------------------
# 1. CARREGAR OS DADOS DO SINAN (O SEU DATA LAKE LOCAL)
# ---------------------------------------------------------
print("A carregar os ficheiros SINAN consolidados...")
pasta_dados = 'dados_siest_parquet'
ficheiros_sinan = [os.path.join(pasta_dados, f) for f in os.listdir(pasta_dados) if f.endswith('.parquet') and 'geolocalizados' not in f]

if not ficheiros_sinan:
    raise FileNotFoundError("Não encontrei os ficheiros do SINAN na pasta 'dados_siest_parquet'.")

df_sinan = pd.concat([pd.read_parquet(f) for f in ficheiros_sinan], ignore_index=True)
print(f"Total de {len(df_sinan)} notificações do SINAN prontas.")

# ---------------------------------------------------------
# 2. CARREGAR E PREPARAR O CSV DO CNES
# ---------------------------------------------------------
print("\nA ler a base de dados do CNES (Ficheiro Local)...")
caminho_cnes_csv = '/content/cnes_estabelecimentos.csv'

# Forçamos a leitura como 'string' para não perder os zeros à esquerda dos códigos
df_cnes = pd.read_csv(caminho_cnes_csv, sep=';', encoding='latin1', dtype=str)

# Filtramos apenas para Campinas (O IBGE às vezes vem com 6 ou 7 dígitos: 350950 ou 3509502)
# O str.startswith garante que apanhamos Campinas de qualquer forma
df_cnes_campinas = df_cnes[df_cnes[COL_MUNICIPIO].str.startswith('350950', na=False)].copy()

print(f"Hospitais/Postos encontrados em Campinas no CNES: {len(df_cnes_campinas)}")

# Isolamos apenas as colunas que importam para o Grafo
df_cnes_campinas = df_cnes_campinas[[COL_CNES, COL_NOME, COL_LAT, COL_LON]]

# Renomeamos para o padrão que vai para o MongoDB
df_cnes_campinas.rename(columns={
    COL_CNES: 'ID_UNIDADE',
    COL_LAT: 'LATITUDE',
    COL_LON: 'LONGITUDE'
}, inplace=True)

# ---------------------------------------------------------
# 3. O CRUZAMENTO ESPACIAL (MERGE)
# ---------------------------------------------------------
print("\nA realizar o cruzamento geoespacial...")

# Limpeza rigorosa das chaves para garantir que o JOIN bate a 100%
df_sinan['ID_UNIDADE'] = df_sinan['ID_UNIDADE'].astype(str).str.replace(r'\.0$', '', regex=True)
df_sinan['ID_UNIDADE'] = df_sinan['ID_UNIDADE'].replace('nan', pd.NA)

df_cnes_campinas['ID_UNIDADE'] = df_cnes_campinas['ID_UNIDADE'].astype(str).str.replace(r'\.0$', '', regex=True)

# O glorioso Left Join
df_siest_geo = pd.merge(df_sinan, df_cnes_campinas, on='ID_UNIDADE', how='left')

# ---------------------------------------------------------
# 4. GUARDAR O FICHEIRO MESTRE
# ---------------------------------------------------------
ficheiro_final = 'dados_siest_parquet/siest_casos_geolocalizados.parquet'
df_siest_geo.to_parquet(ficheiro_final, index=False, compression='snappy')

print(f"\n✅ PIPELINE ETL FECHADO COM SUCESSO! A base de dados mestre está guardada em: {ficheiro_final}")

# Visualização da matriz final
print("\nPrévia de 5 casos geolocalizados:")
colunas_visualizar = ['DT_NOTIFIC', 'ID_UNIDADE', 'NO_FANTASIA', 'LATITUDE', 'LONGITUDE']
print(df_siest_geo[colunas_visualizar].dropna().head())
```
