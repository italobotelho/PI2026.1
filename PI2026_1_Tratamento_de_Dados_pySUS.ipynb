```python
import pandas as pd
from pysus.online_data import IBGE
from pysus.ftp.databases.sim import SIM
from pysus.ftp.databases.sinan import SINAN
```

```python
# 1. Inicializa o SINAN
sinan = SINAN().load()

# 2. Busca os arquivos de Dengue ('DENG') para um ano recente (ex: 2022 ou 2023)
# Baixar múltiplos anos pode demorar um pouco
arquivos_dengue = sinan.get_files('DENG', 2022)

# 3. Faz o download e converte para DataFrame Pandas
df_dengue = sinan.download(arquivos_dengue).to_dataframe()

# 4. Transforma a coluna de Município em número para facilitar o filtro
df_dengue['ID_MUNICIP'] = pd.to_numeric(df_dengue['ID_MUNICIP'], errors='coerce')

# 5. FILTRA PARA CAMPINAS (Código IBGE do município: 350950)
df_campinas = df_dengue[df_dengue['ID_MUNICIP'] == 350950].copy()

# 6. Converte as colunas de Data
df_campinas['DT_NOTIFIC'] = pd.to_datetime(df_campinas['DT_NOTIFIC'], errors='coerce')
df_campinas['DT_SIN_PRI'] = pd.to_datetime(df_campinas['DT_SIN_PRI'], errors='coerce')

# 7. Visualiza os dados principais para o SIEST
colunas_interesse = ['DT_NOTIFIC', 'DT_SIN_PRI', 'ID_UNIDADE', 'CLASSI_FIN']
print(f"Total de casos em Campinas no ano selecionado: {len(df_campinas)}")
df_campinas[colunas_interesse].head()
```

```python
# vasculhar as colunas para encontrar a que se refere a Bairro
colunas_bairro = [col for col in df_campinas.columns if 'BAI' in col]
print(f"Possíveis colunas para Bairro: {colunas_bairro}")
```
